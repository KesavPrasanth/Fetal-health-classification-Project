## BOXPLOTS (CHECKING OUTLIERS)

```python
plt.figure(figsize=(20,10))
sns.boxplot( data = df, orient="h")
plt.show()
```

```python
Q1 = np.percentile(df["percentage_of_time_with_abnormal_long_term_variability"], 25, interpolation = 'midpoint') 
Q2 = np.percentile(df["percentage_of_time_with_abnormal_long_term_variability"], 50, interpolation = 'midpoint') 
Q3 = np.percentile(df["percentage_of_time_with_abnormal_long_term_variability"], 75, interpolation = 'midpoint') 
  
print('Q1 25 percentile of the given data is, ', Q1)
print('Q1 50 percentile of the given data is, ', Q2)
print('Q1 75 percentile of the given data is, ', Q3)
  
IQR = Q3 - Q1 
print('Interquartile range is', IQR)
```

```python
low_lim = Q1 - 1.5 * IQR
up_lim = Q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
```

```python

outlier =[]
for x in df["percentage_of_time_with_abnormal_long_term_variability"]:
    if ((x> up_lim) or (x<low_lim)):
         outlier.append(x)
print(' outlier in the dataset is', len(outlier))
```

```python
# Computing 10th, 90th percentiles and replacing the outliers
tenth_percentile = np.percentile(df["percentage_of_time_with_abnormal_long_term_variability"], 10)
ninetieth_percentile = np.percentile(df["percentage_of_time_with_abnormal_long_term_variability"], 90)
```

```python
maxVal = 38.0
df["percentage_of_time_with_abnormal_long_term_variability"] = df["percentage_of_time_with_abnormal_long_term_variability"].where(df["percentage_of_time_with_abnormal_long_term_variability"] >= maxVal, maxVal) 
minVal = 0.0
df["percentage_of_time_with_abnormal_long_term_variability"] = df["percentage_of_time_with_abnormal_long_term_variability"].where(df["percentage_of_time_with_abnormal_long_term_variability"] <= minVal, minVal)
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df, 0.9)
len(set(corr_features))
```
